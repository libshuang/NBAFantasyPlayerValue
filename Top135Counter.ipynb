```python
df = pd.read_csv('prrankings2020alum.csv')
df
```

```python
df.drop(columns = ['FG%', 'FGA', 'FT%', 'FTA', '3PA' ,'3P%', '3P%_Adj'], inplace = True)
```

```python
df['Top1'] = ''
df['Top3'] = ''
df['Top5'] = ''
```

```python
for j in range(0,len(df)):
    counter = 0
    for i in range(1,11):
        if df.loc[j][i] > 1.96:
            counter += 1
    df['Top1'][j] = counter
```

```python
for j in range(0,len(df)):
    counter = 0
    for i in range(1,11):
        if df.loc[j][i] >= 1.5:
            counter += 1
    df['Top3'][j] = counter
```

```python
for j in range(0,len(df)):
    counter = 0
    for i in range(1,11):
        if df.loc[j][i] >= 1.0:
            counter += 1
    df['Top5'][j] = counter
```
